Group by a column and compute mean, max, and std in one go.

```python
df = pd.DataFrame(
    {
        "City" : np.tile(["Tehran", "Shiraz"], 5),
        "Temperatrue" : np.random.randint(-12, 15, 10)
    }
)

df.groupby(by="City")["Temperatrue"].apply(lambda x :  pd.Series({"mean" : x.mean(), "max" : x.max(), "std" :x.std()}))
```

Create a DataFrame with 3 numeric columns, randomly insert NaNs, fill missing values with column means, and then scale all 
columns between 0 and 1.

# Row Labeler :
 
Create a DataFrame with columns 'score', 'group', and a datetime index. Label each row as 'high' or 'low' based on whether 
the score is above the median within that group.

```python
ts = pd.date_range("2013/1/23",periods=12)

df = pd.DataFrame(
    {
        "score" : np.random.randn(12)*5,
        "group" : np.tile(["A", "B", "C"],4)
    },
    index=ts
)

group_median =df.groupby("group")["score"].median()
df['group_median'] = df['group'].map(group_median)
df['label'] = np.where(df['group_median'] > df['score'], "low" , "high")

df
```

# Rolling Average Filter:

Add a rolling mean column (window size = 3) to a time-indexed DataFrame. Then select only the rows where the rolling mean 
is above 0.5.

```python
df = pd.DataFrame(
    {
        "A" : np.random.randint(0,3,10)
    },
    index=pd.date_range("20200305",periods=10)
)

df["A"] = df["A"].rolling(3).mean()

df[df["A"] > 0.5]
```

# Conditional Map Chain:

Create a column of ages and map them to life stages: 'child', 'teen', 'adult', 'senior', then count how many of each stage 
you have.

```python
df= pd.DataFrame(
    {
        "Ages" : np.random.randint(1,90, 100)
    }
)

def stage(age):
    if age < 12:
        return "child"
    elif  age <18:
        return"teen"
    elif age <65:
        return "adult"
    else:
        return "senior"

# df["Ages"].apply(stage).value_counts()
counts = df["Ages"].map(stage).value_counts()


counts
```
